```python
#import modules
import csv
import pandas as pd
import folium
import folium.plugins as plugins

#map startup coordinates = Mt. Marcy
#map is named mtn_map
mtn_map = folium.Map(location = [44.113000, -073.924000], tiles = 'Stamen Terrain')

#read csv file and convert into a panda dataframe
with open('Adirondack High Peaks Coordinates CSV.csv', 'r') as csv_file:
    data = pd.read_csv(csv_file)    

#creates marker clusters that hold the mountains of different ranges
#Adirondack 46er's
dr_cluster = plugins.MarkerCluster(name = 'Dix Range').add_to(mtn_map)
cr_cluster = plugins.MarkerCluster(name = 'Colvin Range').add_to(mtn_map)
gr_cluster = plugins.MarkerCluster(name = 'Great Range').add_to(mtn_map)
nhp_cluster = plugins.MarkerCluster(name = 'Northern High Peaks').add_to(mtn_map)
str_cluster = plugins.MarkerCluster(name = 'Street Range').add_to(mtn_map)
mm_cluster = plugins.MarkerCluster(name = 'McIntyre Mountains').add_to(mtn_map)
mg_cluster = plugins.MarkerCluster(name = 'Marcy Group').add_to(mtn_map)
wma_cluster = plugins.MarkerCluster(name = 'Whiteface Mountain Area').add_to(mtn_map)
gma_cluster = plugins.MarkerCluster(name = 'Giant Mountain Area').add_to(mtn_map)
sar_cluster = plugins.MarkerCluster(name = 'Santanoni Range').add_to(mtn_map)
sm_cluster = plugins.MarkerCluster(name = 'Seward Mountains').add_to(mtn_map)
#Vermont Ranges
#gmr_cluster = plugins.MarkerCluster(name = 'Green Mountain Range').add_to(mtn_map)
#iterates over the panda dataframe and plot the markers on the map 
for row in data.to_records():
    
    #Determines the color of the marker dependening on the elevation of the mountain
    if row['Elevation'] > 5000:
        color_information = 'red'
        
    elif row['Elevation'] >= 4500:
        color_information = 'blue'
        
    elif row['Elevation'] >= 4000:
        color_information = 'green'
        
    elif row['Elevation'] < 4000:
        color_information = 'lightgreen'
        
    #coordinates of mountain for marker
    coords = (row['Longitude'], row['Latitude'])
    #information that is in the popup
    popup_information = "%s <br> %d ft | %s ft ascent" % (row['Name'], row['Elevation'], row['Ascent of Climb']) 
    #marker creation
    marker = folium.Marker(location = coords, popup = popup_information, icon = folium.Icon(color = color_information))
    
    #plots marker on to map depending on which range it is a part of - reads 'Range' column in .csv
    if row['Range'] == 'Dix Range':
        dr_cluster.add_child(marker)
        
    elif row['Range'] == 'Colvin Range':
        cr_cluster.add_child(marker)
        
    elif row['Range'] == 'Great Range':
        gr_cluster.add_child(marker)
        
    elif row['Range'] == 'Northern High Peaks':
        nhp_cluster.add_child(marker)
        
    elif row['Range'] == 'Street Range':
        str_cluster.add_child(marker)
        
    elif row['Range'] == 'McIntyre Mountains':
        mm_cluster.add_child(marker)
        
    elif row['Range'] == 'Marcy Group':
        mg_cluster.add_child(marker)
        
    elif row['Range'] == 'Whiteface Mountain Area':
        wma_cluster.add_child(marker)
    
    elif row['Range'] == 'Giant Mountain Area':
        gma_cluster.add_child(marker)
        
    elif row['Range'] == 'Santanoni Range':
        sar_cluster.add_child(marker)
        
    elif row['Range'] == 'Seward Mountains':
        sm_cluster.add_child(marker)
        
    #elif row['Range'] == 'Green Mountain Range':
        sm_cluster.add_child(marker)
        
#layer control used to sort map layers 
folium.LayerControl().add_to(mtn_map)

#saves map as html file
mtn_map.save('folium_ADK_mtn_map.html')

#outputs map in IDE
mtn_map
```
